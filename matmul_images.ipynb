```python
img1 = img1.crop((0,50,960,960+50))
```

```python
img1_np = np.array(img1)
img2_np = np.array(img2)
```

```python
tensor_1 = torch.tensor(img1_np,dtype = torch.float32)
tensor_2 = torch.tensor(img2_np,dtype = torch.float32)
```

```python
tensor_1.shape
```

```python
result = tensor_2 @ tensor_1
```

```python
result.cpu().numpy()
```

```python
result = (result - result.min()) / (result.max() - result.min())*255
```
